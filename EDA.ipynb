- Outliers check

```python
# Specify the column you want to check for outliers
column_name = df.ORIGINAL_COMBINED_LOAN_TO_VALUE

# Calculate the z-scores for the column
z_scores = np.abs((column_name - column_name.mean()) / column_name.std())

# Define a threshold for outlier detection
threshold = 3

# Identify outliers
outliers = df[z_scores > threshold]

# Print the outliers
print(outliers)
```

Feature Relationships

```python
#Credit Score vs Original Loan to Value
continuous_variables_1 = ['CREDIT_SCORE', 'ORIGINAL_LOAN_TO_VALUE']
fig,ax = plt.subplots(1,2, figsize = (15,6))
for i,x in enumerate(continuous_variables_1):
    sns.boxplot(x = df.DELINQUENT, y = df[x], ax = ax[i], palette="Set2")
    
#Set custom labels
ax[0].set_xticklabels(['Not Delinquent', 'Delinquent'])
ax[1].set_xticklabels(['Not Delinquent', 'Delinquent'])
```

```python
def exp_num_cols(df, col, title="", fig_size=(12, 7), x_label=""):

    print("Summary statistics:\n")
    print(df[col].describe())
    print("\n\nNumber of missing values: {}".format(df[col].isnull().sum()))

    fig, ax = plt.subplots()
    plt.title(title)
    df.loc[df.DELINQUENT == True, col].plot.hist(bins=50, density=True,
                                                   ax=ax, alpha=0.4, label="Default")
    df.loc[df.DELINQUENT == False, col].plot.hist(bins=50, density=True,
                                                  ax=ax, alpha=0.4, label="No Default")
    plt.xlabel(x_label)
    plt.ylabel("Density Values")
    plt.legend()
    plt.show()

exp_num_cols(df, 'CREDIT_SCORE', title="Distribution of Credit Score Default 
             vs. No Default",
             fig_size=(12, 7), x_label="Credit Score")
```

```python
#Correlation Matrix
df_numeric = df[['CREDIT_SCORE', 'MORTGAGE_INSURANCE_PERCENTAGE', 'ORIGINAL_COMBINED_LOAN_TO_VALUE', 'ORIGINAL_LOAN_TO_VALUE','ORIGINAL_UPB', 'ORIGINAL_INTEREST_RATE', 'PREPAID', 'DELINQUENT']]
df_corr = df_numeric.corr()
sns.heatmap(df_corr, annot=True, fmt = ".2f" )
```
